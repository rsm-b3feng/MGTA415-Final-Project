```python
import matplotlib.pyplot as plt

# Calculate the number of missing values per column
missing_values = df.isnull().sum()

# Plot the missing values
plt.figure(figsize=(12, 6))
missing_values.plot(kind="bar", color="orange")
plt.xlabel("Variables")
plt.ylabel("Number of Missing Values")
plt.title("Frequency of Missing Values per Variable")
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract city names from the location column
df["city"] = df["location"].apply(lambda x: x.split(",")[0] if pd.notnull(x) else "Unknown")

# Count the percentage of users by city
city_counts = df["city"].value_counts(normalize=True) * 100

# Select the top 10 cities
top_cities = city_counts.head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
top_cities.plot(kind="bar", color="blue")
plt.xlabel("City")
plt.ylabel("Percentage of Users")
plt.title("Top 10 Cities by Percentage of Users")
plt.xticks(rotation=45)
plt.show()
```
